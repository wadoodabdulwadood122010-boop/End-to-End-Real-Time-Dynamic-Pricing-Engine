```python
import logging
import pandas as pd
import dagshub
import mlflow
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,  mean_squared_error, r2_score
```

```python
## For local use , not for github
mlflow.set_tracking_uri('https://dagshub.com/wadoodabdulwadood122010/End-to-End-Real-Time-Dynamic-Pricing-Engine.mlflow')
dagshub.init(repo_owner='wadoodabdulwadood122010', repo_name='End-to-End-Real-Time-Dynamic-Pricing-Engine', mlflow=True)
mlflow.set_experiment('baseline experiment')
```

```python
df = pd.read_csv(r'C:\mlops\End-to-End-Real-Time-Dynamic-Pricing-Engine\notebooks\data\processed\processed.csv')
```

```python
df  = df.drop(columns = ['Unnamed: 0'])
```

```python
df.head()
```

```python
x  = df.drop(columns = ['Historical_Cost_of_Ride'])
y = df['Historical_Cost_of_Ride']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.25,random_state=42)
```

```python
import mlflow
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

with mlflow.start_run(run_name='Run 1'):
    try:
        print('Logging params ...')
        mlflow.log_params({'test_size': 0.25, 'random_state': 42})

        print('Loading model ...')
        model = LinearRegression() # Fits continuous data perfectly

        print('Training model ...')
        model.fit(x_train, y_train)
        y_pred = model.predict(x_test)

        print('Logging metrics ...')
        # Valid metrics for continuous data
        mse = mean_squared_error(y_test, y_pred)
        r2 = r2_score(y_test, y_pred)
        
        print(f'   Mean Squared Error: {mse}')
        print(f'   R2 Score: {r2}')
        
        mlflow.log_metric('mse', mse)
        mlflow.log_metric('r2_score', r2)


        mlflow.sklearn.log_model(model, name="my_model")
    except Exception as e:
        print(f"An error occurred: {e}")
```

```python
import mlflow
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score

with mlflow.start_run(run_name='Run 2'):
    try:
        print('Logging params ...')
        mlflow.log_params({'test_size': 0.25, 'random_state': 42, 'n_estimators':100})

        print('Loading model ...')
        model = RandomForestRegressor(n_estimators=100, random_state=42)

        print('Training model ...')
        model.fit(x_train, y_train)
        y_pred = model.predict(x_test)

        print('Logging metrics ...')
        # Valid metrics for continuous data
        mse = mean_squared_error(y_test, y_pred)
        r2 = r2_score(y_test, y_pred)
        score = model.score(x_test, y_test)
        
        print(f'   Mean Squared Error: {mse}')
        print(f'   R2 Score: {r2}')
        print(f'   score:', {score})
        
        mlflow.log_metric('mse', mse)
        mlflow.log_metric('r2_score', r2)
        mlflow.log_metric('score', score)


        mlflow.sklearn.log_model(model, name="my_model")
    except Exception as e:
        print(f"An error occurred: {e}")
```
